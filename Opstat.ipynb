Fibonacci $f(x) = -\frac{1}{(x-1)^2} \left(\log(x) - \frac{2(x-1)}{x+1}\right)$

```python
import sympy as sp

# Mendefinisikan simbol x
x = sp.symbols('x')

# Mendefinisikan fungsi
f = -1/(x-1)**2 * (sp.log(x) - 2*(x-1)/(x+1))

# Menghitung turunan pertama
f_prime = sp.diff(f, x)

# Menghitung turunan kedua
f_double_prime = sp.diff(f_prime, x)

# Menampilkan turunan kedua
print(f"Turunan kedua: {f_double_prime}")

# Evaluasi turunan kedua pada x = 2, x = 3, dan x = 4
for val in [2, 3, 4]:
    result = f_double_prime.subs(x, val)
    print(f"Nilai turunan kedua pada x = {val} adalah {result}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Mendefinisikan fungsi
def f(x):
    return -1/(x-1)**2 * (np.log(x) - 2*(x-1)/(x+1))

# Membuat array x dari 1.0 hingga 5.0
x_values = np.linspace(1.01, 5.0, 400)  # Menghindari pembagian oleh 0

# Menghitung nilai y (f(x)) untuk setiap nilai x
y_values = f(x_values)

# Membuat plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='f(x) = -1/(x-1)^2 (log x - 2(x-1)/(x+1))')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot Fungsi f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
