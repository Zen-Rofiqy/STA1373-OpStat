 Gunakan algoritma Fibonacci untuk mencari titik optimum dari fungsi berikut, 
 
 jika diketahui titik optimum terletak pada selang (1.5; 4.5)

Fibonacci $f(x) = -\frac{1}{(x-1)^2} \left(\log(x) - \frac{2(x-1)}{x+1}\right)$
### Part 1

1. Iterasi yang dibutuhkan sehingga mencapai nilai optimum dengan toleransi 0 adalah 100
2. Nilai titik yang mengoptimumkan persamaan tersebut adalah 2.1887
3. Berikut merupakan pernyataan yang benar adalah : Kurva yang terbentuk pada selang (1.0;5.0) merupakan kurva cekung ke atas
4.  Nilai fungsi yang mengoptimumkan persamaan tersebut adalah -0.0267


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -1/((x-1)**2) * (np.log(x) - 2*(x-1)/(x+1))

def f_prime(x):
    return 1/((x-1)**3) * (2*(x-1)/(x+1) - np.log(x)) - 1/((x-1)**2) * (1/x - 2/(x+1))

def f_double_prime(x):
    return -3/((x-1)**4) * (2*(x-1)/(x+1) - np.log(x)) + 2/((x-1)**3) * (1/x - 2/(x+1)) + 1/((x-1)**2) * (1/(x**2) + 2/((x+1)**2))

def fibonacci_optimizer(f, a, b, tol=1e-6, n=100):
    phi = (1 + np.sqrt(5)) / 2
    x1 = b - (b - a) / phi
    x2 = a + (b - a) / phi
    f1, f2 = f(x1), f(x2)
    iteration_count = 0
    for i in range(n):
        iteration_count += 1
        if f1 > f2:
            a = x1
            x1 = x2
            f1 = f2
            x2 = a + (b - a) / phi
            f2 = f(x2)
        else:
            b = x2
            x2 = x1
            f2 = f1
            x1 = b - (b - a) / phi
            f1 = f(x1)
        if abs(b - a) < tol:
            break
    print(f"Jumlah iterasi yang dibutuhkan dengan toleransi {tol}: {iteration_count}")
    return (a + b) / 2

# Mengevaluasi turunan kedua pada selang (1.0; 5.0)
x = np.linspace(1.0, 5.0, 100)
f_double_prime_values = [f_double_prime(xi) for xi in x]

# Jika semua nilai turunan kedua positif, maka kurva cekung ke atas
is_convex_up = all(value > 0 for value in f_double_prime_values)
print(f"Kurva pada selang (1.0; 5.0) {'cekung ke atas' if not is_convex_up else 'tidak cekung ke atas'}")

# Mencari titik optimum dengan algoritma Fibonacci
optimal_x = fibonacci_optimizer(f, 1.5, 4.5, tol=0)
print(f"Titik optimum pada selang (1.5; 4.5): x = {optimal_x:.6f}, f(x) = {f(optimal_x):.6f}")

# Memplot fungsi
x_plot = np.linspace(1.0, 5.0, 1000)
y_plot = [f(xi) for xi in x_plot]
plt.plot(x_plot, y_plot)
plt.axvline(optimal_x, color='r', linestyle='--', label=f'Titik optimum: x = {optimal_x:.6f}')
plt.axhline(0, color='k', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot Fungsi Fibonacci')
plt.legend()
plt.show()
```

```python
import numpy as np

def f(x):
    x1, x2 = x
    return x1**2 + x2**2 - 2*x1 - 4*x2 + 5

def differential_evolution(func, bounds, max_iter=1000, pop_size=50, mutation_factor=0.8, crossover_prob=0.7):
    dimensions = len(bounds)
    population = np.random.uniform(bounds[:, 0], bounds[:, 1], size=(pop_size, dimensions))
    fitness = np.apply_along_axis(func, 1, population)
    best_idx = np.argmin(fitness)

    for _ in range(max_iter):
        for i in range(pop_size):
            indices = [idx for idx in range(pop_size) if idx != i]
            a, b, c = population[np.random.choice(indices, 3, replace=False)]

            mutant = a + mutation_factor * (b - c)
            mutant = np.clip(mutant, bounds[:, 0], bounds[:, 1])

            trial = np.copy(population[i])
            crossover = np.random.rand(dimensions) < crossover_prob
            trial[crossover] = mutant[crossover]
            trial_fitness = func(trial)

            if trial_fitness < fitness[i]:
                population[i] = trial
                fitness[i] = trial_fitness
                if trial_fitness < fitness[best_idx]:
                    best_idx = i

    return population[best_idx]

# Batasan nilai x1 dan x2
x1_min, x1_max = -2, 2
x2_min, x2_max = 1, 3
bounds = np.array([(x1_min, x1_max), (x2_min, x2_max)])

# Menjalankan Differential Evolution
min_point = differential_evolution(f, bounds)

print("Titik minimum yang ditemukan:", min_point)
```

```python
import numpy as np

def simplex_method(f, x1, x2, x3, eps=0.2, trace=True, max_iter=100, alpha=0.1, beta=0.1, gamma=0.1):
    if trace:
        print("\nTRACING\n")
    i = 0
    d = np.array([f(*x1), f(*x2), f(*x3)])  # Memperbaiki cara pemanggilan fungsi f
    Q = float('inf')
    
    while Q > eps and i < max_iter:
        nxl = np.argmin(d)
        nxh = np.argmax(d)
        
        xl, xh = np.array([x1, x2, x3])[nxl], np.array([x1, x2, x3])[nxh]  # Mengonversi list menjadi array numpy
        x0 = 0.5 * (xl + xh)
        
        xr = (1 + alpha) * x0 - alpha * xh
        
        if f(*xr) < f(*xl):  # Memperbaiki cara pemanggilan fungsi f
            xe = gamma * xr + (1 - gamma) * x0
            if f(*xe) < f(*xl):  # Memperbaiki cara pemanggilan fungsi f
                xh = xe
        
        if f(*xr) > f(*x0) and f(*xl) > f(*x0) and f(*xr) < f(*xh):  # Memperbaiki cara pemanggilan fungsi f
            xh = xr
            xc = beta * xh + (1 - beta) * x0
            if f(*xc) < min(f(*xh), f(*xr)):  # Memperbaiki cara pemanggilan fungsi f
                xh = xc
            else:
                xl = (xl + x0) / 2
                xh = (xh + x0) / 2
        
        x1, x2, x3 = sorted([xl, x0, xh], key=lambda x: f(*x))  # Memperbaiki cara pemanggilan fungsi f
        
        d = np.array([f(*x1), f(*x2), f(*x3)])  # Memperbaiki cara pemanggilan fungsi f
        Q = np.sqrt(((f(*x1) - f(*x0))**2 + (f(*x2) - f(*x0))**2 + (f(*x3) - f(*x0))**2) / 3)  # Memperbaiki cara pemanggilan fungsi f
        
        if trace:
            print("------ Iter:", i, "-----")
            print("x1:", x1)
            print("x2:", x2)
            print("x3:", x3)
            print("Q:", round(Q, 2), "\n")
        
        i += 1
    
    print("\n")
    
    if i == max_iter:
        return {"optim": x0, "f.optim": f(*x0), "warning": f"stop.iter sama dengan max.iter yaitu pada iterasi ke-{i}", "Q": Q}
    else:
        return {"optim": x0, "f.optim": f(*x0), "stop.iter": i, "Q": Q}

def f(x1, x2):
    return x1**4 - 2*x1**2*x2 + x1**2 + x2**2 + 2*x1 + 1

x1 = [4, 4]
x2 = [5, 4]
x3 = [4, 5]

result = simplex_method(f, x1, x2, x3, eps=0.2, trace=True, max_iter=100, alpha=1, beta=0.5, gamma=2)
print(result)
```
